```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
sampling_frequency = 100
```

```python
# Samples per period
T = 1/sampling_frequency
```

```python
t_min = 0.0
t_max = 5.0
```

```python
t = np.arange(t_min, t_max, T)
```

```python
t.shape
```

```python
# Signal frequencies
f1 = 5
f2 = 10
```

```python
# Get waves
y1 = np.sin(2*np.pi*f1*t)
y2 = np.sin(2*np.pi*f2*t)
```

```python
sum_wave = y1 + y2
```

```python
# Compute FT of sum signals
ft = np.fft.fft(sum_wave) / len(sum_wave)
ft = abs(ft[range(int(len(sum_wave)/2))])
```

```python
# get frequency domain
n_outputs = len(sum_wave)
values = np.arange(n_outputs//2)
time_period = n_outputs/sampling_frequency
freq = values/time_period
```

```python
fig, axs = plt.subplots(4, figsize=(8, 12))

axs[0].plot(t, y1)
axs[0].grid(True, which='both')
axs[0].axhline(y=0, color='k')
axs[0].set(ylabel='y1 {} Hz'.format(f1))

axs[1].plot(t, y2)
axs[1].grid(True, which='both')
axs[1].axhline(y=0, color='k')
axs[1].set(ylabel='y2 {} Hz'.format(f2))

axs[2].plot(t, sum_wave)
axs[2].grid(True, which='both')
axs[2].axhline(y=0, color='k')
axs[2].set(ylabel='Sum of waves')

axs[3].plot(freq, ft, 'tab:red')
axs[3].grid(True, which='both')
axs[3].axhline(y=0, color='k')
axs[3].set(ylabel='FT')

plt.show()
```

Notice above that sum of two consines and its FT creates two spikes 

## Sum many sine waves with one very high frequency component

```python
y1 = np.sin(2*np.pi*2*t)
y2 = np.sin(2*np.pi*5*t)
y3 = np.sin(2*np.pi*8*t)
y4 = np.sin(2*np.pi*10*t)
y5 = np.sin(2*np.pi*32*t) # high freq
```

```python
sum_wave = y1 + y2 + y3 + y4 + y5
```

```python
ft = np.fft.fft(sum_wave) / len(sum_wave)
ft = abs(ft[range(int(len(sum_wave)/2))])
```

```python
# get frequency domain
n_outputs = len(sum_wave)
values = np.arange(n_outputs//2)
time_period = n_outputs/sampling_frequency
freq = values/time_period
```

```python
fig, axs = plt.subplots(7, figsize=(8, 12))

axs[0].plot(t, y1)
axs[0].grid(True, which='both')
axs[0].axhline(y=0, color='k')

axs[1].plot(t, y2)
axs[1].grid(True, which='both')
axs[1].axhline(y=0, color='k')

axs[2].plot(t, y3)
axs[2].grid(True, which='both')
axs[2].axhline(y=0, color='k')

axs[3].plot(t, y4)
axs[3].grid(True, which='both')
axs[3].axhline(y=0, color='k')

axs[4].plot(t, y5)
axs[4].grid(True, which='both')
axs[4].axhline(y=0, color='k')


axs[5].plot(t, sum_wave)
axs[5].grid(True, which='both')
axs[5].axhline(y=0, color='k')
axs[5].set(ylabel='Sum of waves')

axs[6].plot(freq, ft, 'tab:red')
axs[6].grid(True, which='both')
axs[6].axhline(y=0, color='k')
axs[6].set(ylabel='FT')

plt.show()
```
