```python
import copy
import cv2
import matplotlib.pyplot as plt
import my_img_ops as im
import numpy as np
```

```python
# Read in images
retina_img = cv2.imread('retina.jpeg', 0)
seeds_img = cv2.imread('seeds.jpg', 0)
aorta_img = cv2.imread('aortic_diagram.png', 0)
```

```python
# Change gamma value of image
c = 1
epsilon = 0
gamma = 0.5
res1 = im.change_gamma(c, epsilon, gamma, retina_img, True)
```

```python
res1 = res1 * 255
plt.imshow(res1, cmap='gray', vmin=0, vmax=255)
```

```python
# Contrast stretch
res2 = copy.deepcopy(seeds_img)
res3 = copy.deepcopy(seeds_img)

# Case 1: set min intensity to 0 and max intensity to 255
r_min = np.min(res2)
r_max = np.max(res2)
res2 = im.contrast_stretch(r_min, 0, r_max, 255, res2)

# Case 2: make binary image to highlight pixels with average intensity
avg = int(np.average(res3))
res3 = im.contrast_stretch(avg, 0, avg, 255, res3)
```

```python
plt.imshow(seeds_img, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imshow(res2, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imshow(res3, cmap='gray', vmin=0, vmax=255)
```

```python
# Get image negative
res4 = im.get_negative(aorta_img)
```

```python
plt.imshow(aorta_img, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imshow(res4, cmap='gray', vmin=0, vmax=255)
```

```python
# Case: intensities outside of [150,200] are reduced to 0
res5 = im.intensity_level_slice1(res5, 150, 200, 225, 0)

# Case: intensities outside of [150,200] are left alone
res6 = im.intensity_level_slice2(res6, 150, 200, 255)
```
