```python
# Imports
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in image
filename = 'brussels-airport-runway.jpg'
img = cv2.imread(filename)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
```

```python
# Get edge image (via Canny edge detection)

# Smooth frame using Gaussian filter
sigma = 1
k = 3*sigma
if k%2 == 0:
    k += 1

blur_img = cv2.GaussianBlur(src=img, ksize=(k,k), sigmaX=sigma, sigmaY=sigma)

# Get Canny edges
low_threshold = 255
high_threshold = 255
canny_img = cv2.Canny(blur_img, low_threshold, high_threshold)
```

```python
plt.imshow(canny_img, cmap='gray')
```
