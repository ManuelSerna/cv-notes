```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

import spatial_filtering_funcs as fs
import SpatialFilter as filters
```

```python
img = cv2.imread('sudoku.jpg', 0)
#img = cv2.imread('patterns.png', 0)
```

```python
kernels = filters.SpatialFilter()
```

```python
# Function: plot two original image next to processed image
def plot_imgs(img1=None, title1='Original', img2=None, title2='My Filtered', img3=None, title3='OpenCV Filtered'):
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 4))
    ax = axes.ravel()

    # Original
    ax[0].imshow(img1, cmap='gray')
    ax[0].set_title(title1)

    # My filtered
    ax[1].imshow(img2, cmap='gray')
    ax[1].set_title(title2)
    
    # OpenCV's filtered
    ax[2].imshow(img3, cmap='gray')
    ax[2].set_title(title3)

    plt.tight_layout()
    plt.show()
```

```python
# Apply sobel x filter
kernel = kernels.sobelx
my_sobelx_res = fs.convolution_2d(image=img, w=kernel, flip_filter=False) # my own
opencv_sobelx_res = cv2.Sobel(src=img, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3) # opencv
```

```python
plot_imgs(
    img1=img, 
    img2=my_sobelx_res, 
    title2='My Sobel X', 
    img3=opencv_sobelx_res, 
    title3='OpenCV Sobel X'
)
```

```python
# Apply sobel y filter
kernel = kernels.sobely
my_sobely_res = fs.convolution_2d(img, kernel, flip_filter=False) # my own
opencv_sobely_res = cv2.Sobel(src=img, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3) # opencv
```

```python
plot_imgs(
    img1=img, 
    img2=my_sobely_res, 
    title2='My Sobel Y', 
    img3=opencv_sobely_res, 
    title3='OpenCV Sobel Y'
)
```

```python
# Apply laplacian filter
kernel = kernels.laplacian1
my_lap_res = fs.convolution_2d(img, kernel, flip_filter=False)
opencv_lap_res = cv2.Laplacian(img, cv2.CV_16S, 3)
```

```python
plot_imgs(
    img1=img, 
    img2=my_lap_res, 
    title2='My Laplacian', 
    img3=opencv_lap_res, 
    title3='OpenCV Laplacian'
)
```

```python
# Apply gaussian filter
kernel = kernels.gaussian
my_gauss_res = fs.convolution_2d(img, kernel, flip_filter=False)
opencv_gauss_res = cv2.GaussianBlur(img,(3,3),0)
```

```python
plot_imgs(
    img1=img, 
    img2=my_gauss_res, 
    title2='My Gaussian', 
    img3=opencv_gauss_res, 
    title3='OpenCV Gaussian'
)
```
