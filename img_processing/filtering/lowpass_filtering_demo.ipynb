```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

import spatial_filtering_funcs as fs
import SpatialFilter as filters
```

```python
# Function: plot two images side-by-side
def plot_imgs(img1=None, title1='', img2=None, title2=''):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
    ax = axes.ravel()

    ax[0].imshow(img1, cmap='gray')
    ax[0].set_title(title1)
    ax[1].imshow(img2, cmap='gray')
    ax[1].set_title(title2)    

    plt.tight_layout()
    plt.show()
```

```python
img1 = cv2.imread('patterns.png', 0)
```

```python
# Use box filter
size = 7
box_filter = fs.get_box_filter(size=size)
```

```python
img_box_smoothed = fs.convolution_2d(
    image=img1, 
    w=box_filter, 
    scale=False, 
    flip_filter=True
)
```

```python
plot_imgs(
    img1, 
    'Original', 
    img_box_smoothed, 
    'Smoothed (Box) ({}x{})'.format(size, size)
)
```

```python
# Use Gaussian filter
size = 7
gauss_filter = fs.get_gaussian_filter(
    k=size,
    gen_const=1,
    stdev=1
)
```

```python
img_box_smoothed = fs.convolution_2d(
    image=img1, 
    w=gauss_filter, 
    scale=False, 
    flip_filter=True
)
```

```python
plot_imgs(
    img1, 
    'Original', 
    img_box_smoothed, 
    'Smoothed (Gauss) ({}x{})'.format(size, size)
)
```

```python
# Apply median filter
img2 = cv2.imread('xray_circuit.png', 0)
```

```python
size = 7
median_filtered = fs.median_filter(img2, size)
```

```python
plot_imgs(
    img2, 
    'Original', 
    median_filtered, 
    'Median filtered ({}x{})'.format(size, size)
)
```

```python
gsize = 19
gauss_filter_2 = fs.get_gaussian_filter(
    k=gsize,
    gen_const=1,
    stdev=3
)
```

```python
gauss_filtered = fs.convolution_2d(
    image=img2, 
    w=gauss_filter_2, 
    scale=False, 
    flip_filter=True
)
```

```python
plot_imgs(
    median_filtered, 
    'Median filtered ({}x{})'.format(size, size),
    gauss_filtered,
    'Gauss filtered ({}x{}), stdev={}'.format(gsize, gsize, 3)
)
```
